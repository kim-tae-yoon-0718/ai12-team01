# Kaggle Credential Setup\n\nCopy `.env.example` to `.env`, fill in your own Kaggle credentials, then run the cell below. Do not commit `.env` or `kaggle.json`.\n

```python
import json\nimport os\nfrom pathlib import Path\n\n\ndef load_dotenv(path=Path(".env")):\n    """Load simple KEY=VALUE pairs without requiring python-dotenv."""\n    path = Path(path)\n    if not path.exists():\n        return\n    for raw_line in path.read_text().splitlines():\n        line = raw_line.strip()\n        if not line or line.startswith("#") or "=" not in line:\n            continue\n        key, value = line.split("=", 1)\n        key = key.strip()\n        value = value.strip().strip('"').strip("'")\n        if key and value and not os.environ.get(key):\n            os.environ[key] = value\n\n\nload_dotenv()\n\nusername = os.environ.get("KAGGLE_USERNAME")\nkey = os.environ.get("KAGGLE_KEY")\n\nif not username or not key:\n    raise RuntimeError(\n        "Set KAGGLE_USERNAME and KAGGLE_KEY in your environment, "\n        "or copy .env.example to .env and fill in your own values."\n    )\n\nkaggle_dir = Path.home() / ".kaggle"\nkaggle_dir.mkdir(parents=True, exist_ok=True)\ncredential_path = kaggle_dir / "kaggle.json"\ncredential_path.write_text(json.dumps({"username": username, "key": key}))\ncredential_path.chmod(0o600)\n\nprint(f"Kaggle credentials written to {credential_path}")\n
```
